### Python virtual environments

Τα virtual environments της Python είναι ένα θεμελιώδες εργαλείο για τη διαχείριση των εξαρτήσεων και των εκδόσεων του interpreter της γλώσσας που σχετίζονται με κάποιο συγκεκριμένο project, χωρίς να επηρεάζεται η gloabl εγκατάσταση της Python. Ένα σύνηθες φαινόμενο που πρέπει να αντιμετωπιστεί κατά την εργασία σε διαφορετικά python projects είναι το γεγονός ότι κάθε project δύναται να έχει διαφορετικές απαιτήσεις όσον αφορά στις βιβλιοθήκες και τις εκδόσεις αυτών που είναι απαραίτητες. Τα virtual environments επιτρέπουν στους προγραμματιστές να δημιουργούν απομονωμένους χώρους στα συστήματά τους για κάθε project, διασφαλίζοντας ότι το καθένα μπορεί να έχει το δικό του σύνολο εξαρτήσεων και ρυθμίσεων. Αυτή η απομόνωση αποτρέπει τις συγκρούσεις μεταξύ των απαιτήσεων των projects, επιτρέποντας ομαλότερες ροές ανάπτυξης και ευκολότερη διαχείριση εξαρτήσεων.

Η σημασία των virtual environments εκτείνεται πέρα από τη διαχείριση εξαρτήσεων. Διαδραματίζουν επίσης κρίσιμο ρόλο στην αναπαραγωγή των περιβαλλόντων ανάπτυξης προγραμμάτων έτσι ώστε να εξασφαλιστεί η συνοχή όσον αφορά στην ανάπτυξη, τις δοκιμές και της παραγωγής λογισμικού. Αυτή η σταθερότητα βοηθά στην ελαχιστοποίηση του προβλήματος "δουλεύει στο δικό μου μηχάνημα", μια κοινή πρόκληση στην ανάπτυξη λογισμικού, όπου ο κώδικας συμπεριφέρεται διαφορετικά ανάλογα με το λειτουργικό σύστημα του εκάστοτε χρήστη. Επιπλέον, τα εικονικά περιβάλλοντα διευκολύνουν τη συνεργασία μεταξύ των μελών της ομάδας και συμβάλλουν σε πιο αξιόπιστες διαδικασίες ανάπτυξης. Χρησιμοποιώντας ένα αρχείο **requirements.txt** ή παρόμοιους μηχανισμούς για τον καθορισμό των εξαρτήσεων ενός project, οι προγραμματιστές μπορούν να διασφαλίσουν ότι όλα τα μέλη της ομάδας και τελικά το περιβάλλον παραγωγής χρησιμοποιούν ακριβώς τις ίδιες εκδόσεις βιβλιοθηκών, μειώνοντας τις πιθανότητες απροσδόκητης συμπεριφοράς ή σφαλμάτων που είναι δύσκολο να εντοπιστούν.

Μπορείτε να δημιουργήσετε ένα python virtual environment ανοίγοντας ένα terminal και δίνοντας την ακόλουθη εντολή: **python -m venv venv**. Αξίζει να σημειωθεί ότι σε ορισμένα συστήματα linux η εντολή θα πρέπει να είναι **python3 -m venv venv**

Αφού δημιουργήσετε το virtual environment, θα πρέπει να το ενεργοποιήσετε για να μπορέσετε να το χρησιμοποιήσετε. Αυτό μπορείτε να το κάνετε μέσω της ακόλουθης εντολής: **source venv/bin/activate**. Αν θέλετε να ενεργοποιήσετε το venv σας στα windows η εντολή που πρέπει να χρησιμοποιήσετε είναι η εξής: **venv\Scripts\activate**. Ανεξάρτητα από το λειτουργικό σύστημα, βεβαιωθείτε ότι έχετε ανοίξει το terminal στον ίδιο φάκελο όπου έχει δημιουργηθεί το venv σας. Στην Python, όταν αναθέτουμε μια μεταβλητή σε μια άλλη, ειδικά στο πλαίσιο πολύπλοκων δομών δεδομένων όπως λίστες, λεξικά κ.α, δεν δημιουργείτε ένα νέο, ανεξάρτητο αντίγραφο του αρχικού αντικειμένου. Αντίθετα, δημιουργείτε μια νέα αναφορά στο ίδιο αντικείμενο. Αυτό σημαίνει ότι οποιεσδήποτε αλλαγές που γίνονται στα δεδομένα μέσω μιας αναφοράς θα επηρεάσουν το αρχικό αντικείμενο και όλες τις άλλες αναφορές σε αυτό. Αυτή η συμπεριφορά είναι αυτό που συχνά αναφέρεται ως "σφάλμα pass by reference".

Αυτό το σφάλμα πρέπει να αντιμετωπίζεται άμεσα για τους παρακάτω λόγους:



     1) Απρόβλεπτες παρενέργειες: Αν τροποποιήσετε ακούσια το αρχικό αντικείμενο, ενώ σκοπεύατε να αλλάξετε μόνο ένα αντίγραφο, αυτό μπορεί να οδηγήσει σε σφάλματα που είναι δύσκολο να εντοπιστούν. Για παράδειγμα, αν το matrix_2 προορίζεται να είναι μια τροποποιημένη έκδοση του matrix χωρίς να επηρεάζει το matrix, η απευθείας ανάθεση matrix_2 = matrix και στη συνέχεια η αλλαγή του matrix_2 θα αλλάξει και το matrix.

     2) Ακεραιότητα δεδομένων: Για εφαρμογές όπου τα αρχικά δεδομένα πρέπει να παραμείνουν αμετάβλητα για μελλοντική αναφορά ή λειτουργίες, η ακούσια τροποποίηση μέσω μιας αναφοράς μπορεί να θέσει σε κίνδυνο την ακεραιότητα των δεδομένων σας.

````python
# Ένα matrix  μπορεί να δημιουργηθεί μέσω πολλών λιστών
matrix = [
         [1,2,3],
         [4,5,6]
         ]

print(matrix[1][-2]) 

# σφάλμα pass-by-reference
matrix_2 = matrix
matrix_2[1][-2] = 10
print(matrix[1][-2]) 
print(matrix)
````

Για να διαχειριστείτε ή να αποφύγετε αυτό το πρόβλημα, μπορείτε να δημιουργήσετε αντίγραφα των αντικειμένων όταν χρειάζεται να τα τροποποιήσετε ανεξάρτητα από τα πρωτότυπά τους. Για ρηχά αντίγραφα (ένα νέο container με αναφορές στα ίδια αντικείμενα), μπορείτε να χρησιμοποιήσετε μεθόδους όπως η list.copy() ή η συνάρτηση copy(). Για βαθιά αντίγραφα (ένα νέο container και νέα αντίγραφα όλων των περιεχόμενων αντικειμένων), χρησιμοποιήστε τη συνάρτηση copy.deepcopy().

````python
import copy

matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

# Δημιουργία deep copy του 'matrix' έτσι ώστε το 'matrix_2' να είναι ένα εντελώς ξεχωριστό αντικείμενο.
matrix_2 = copy.deepcopy(matrix)

# Τώρα η τροποποίηση του 'matrix_2' δεν θα επηρεάσει το 'matrix'.
matrix_2[1][-2] = 10

print(matrix[1][-2])
print(matrix)
print(matrix_2) 
````

##### Sets
- Είναι ακόμα μια δομή πινάκων της Python
- Αντί για <code>**squared brackets**[]</code> χρησιμοποιούμε <code>**curly brackets**{}</code> και διαχωρίζουμε πάλι τα στοιχεία με κόμματα.
- Η διαφορά τους σε σχέση με τις λίστες είναι ότι δεν μπορούν να αλλάξουν τα στοιχεία τους.
- Δεν επιτρέπουν το ίδιο στοιχείο πάνω από μια φορά.
- Τα στοιχεία δεν διατηρούν κάποια συγκεκριμένη σειρά.
- Είναι δηλαδή λίστες μοναδικών στοιχείων.

````python
pantry = set(("spam", "bacon", "sausage"))
print(pantry)
pantry = {"spam", "bacon", "sausage"}
print(pantry)
````

````python
for x in pantry:
    print(x)
print()
print("bacon" in pantry)
````

````python
# Προσθήκη επιπλέον στοιχείων στο set
pantry.add("ham")
print(pantry)
````

````python
# Μπορούμε επίσης να προσθέσουμε πολλά στοιχεία ταυτόχρονα
pantry.update(["milk", "cheese", "yogurt"])
print(pantry)
````

Αξίζει να σημειωθεί ότι δεν υπάρχει συνάρτηση connect_neighborhood() στην έκδοση του igraph της python. Αλλά αυτό δεν είναι πρόβλημα, μπορούμε να δημιουργήσουμε τη δική μας συνάρτηση που θα παρέχει την ίδια χρησιμότητα με την αντίστοιχη της R.

````python
ring_graph = Graph.Ring(20)

def connect_neighborhood(graph, distance):
    vertex_count = len(graph.vs)
    edges_to_add = []
    for v in range(vertex_count):
        for d in range(1, distance + 1):
            neighbor = (v + d) % vertex_count
            if not graph.are_connected(v, neighbor):
                edges_to_add.append((v, neighbor))
    graph.add_edges(edges_to_add)

# Σύνδεση γειτονιάς κόμβων με απόσταση 8
connect_neighborhood(ring_graph, 8)

visual_style = {
    "vertex_size": 8,
    "vertex_label": None,
    "bbox": (400, 400),
    "margin": 20 
}

plot(ring_graph, **visual_style)
````

Μπορούμε επίσης να υπολογίσουμε τα hubs και τα authorities ενός γράφου, όπως ακριβώς κάνουμε στην R. Για να το κάνουμε αυτό όμως και να οπτικοποιήσουμε το αποτέλεσμα θα πρέπει επίσης να χρησιμοποιήσουμε τη βιβλιοθήκη networkx

````python
# Μετατροπή αντικειμένου igraph σε networkx για οπτικοποίηση
def igraph_to_networkx(ig_graph):
    g = nx.DiGraph() if ig_graph.is_directed() else nx.Graph()
    g.add_nodes_from(range(ig_graph.vcount()))
    g.add_edges_from([(e.source, e.target) for e in ig_graph.es])
    return g

nx_graph = igraph_to_networkx(graph_dir)

# Υπολογισμός hubs και authorities με χρήση igraph
hubs, authorities = graph_dir.hub_score(), graph_dir.authority_score()

# Κανονικοποίηση των scores
hub_sizes = np.array(hubs) * 1000  # Adjust size scaling as necessary
auth_sizes = np.array(authorities) * 1000  # Adjust size scaling as necessary

# Οπτικοποίηση αποτελεσμάτων
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Hubs
nx.draw(nx_graph, ax=axes[0], node_size=hub_sizes, with_labels=False, arrows=True, node_color='skyblue')
axes[0].set_title('Hubs')

# Authorities
nx.draw(nx_graph, ax=axes[1], node_size=auth_sizes, with_labels=False, arrows=True, node_color='lightgreen')
axes[1].set_title('Authorities')

plt.suptitle('Hub and Authority Scores')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to make room for the main title
plt.show()
````

### Υποομάδες, κοινότητες, κλίκες και διάσχιση γράφου

Μπορούμε επίσης να εκτελέσουμε BFS καθώς και DFS στο γράφο και να οπτικοποιήσουμε το αποτέλεσμα

````python
# Εκτέλεση και οπτικοποίηση DFS
dfs_result = undir_graph.dfs(3, mode="all")

node_colors = ["lightblue"] * undir_graph.vcount()
edge_colors = ["grey"] * undir_graph.ecount()


for child, parent in zip(dfs_result[0], dfs_result[1]):
    if parent != -1:
        edge_id = undir_graph.get_eid(parent, child)
        edge_colors[edge_id] = "red"

node_colors[3] = "orange"

undir_graph.vs["color"] = node_colors
undir_graph.es["color"] = edge_colors

visual_style = {
    "vertex_label": list(range(undir_graph.vcount())),
    "edge_arrow_size": 0.5,
    "bbox": (400, 400),
    "margin": 20,
    "main": "DFS Traversal Path Highlighted"
}

plot(undir_graph, **visual_style)
````

````python
node_colors = ["lightblue"] * undir_graph.vcount()
edge_colors = ["grey"] * undir_graph.ecount()

# Εκτέλεση διάσχισης με BFS
bfs_result = undir_graph.bfs(3, mode="all")

node_colors[3] = "orange"

predecessors = bfs_result[2]

for i, parent in enumerate(predecessors):
    if parent >= 0:
        edge_id = undir_graph.get_eid(parent, i, directed=False, error=False)
        if edge_id != -1:
            edge_colors[edge_id] = "red"
            node_colors[i] = "lightblue"

undir_graph.vs["color"] = node_colors
undir_graph.es["color"] = edge_colors

visual_style = {
    "vertex_label": list(range(undir_graph.vcount())),
    "edge_arrow_size": 0.5,
    "bbox": (300, 300),
    "margin": 20,
    "vertex_size": 20
}

# Οπτικοποίηση διάσχισης με BFS
plot(undir_graph, **visual_style)
````
